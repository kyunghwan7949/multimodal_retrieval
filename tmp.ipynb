```python
import numpy as np

mask = np.hstack([
    np.zeros(196 - 147), # UNMASK
    np.ones(147),                     # MASK
])
tmp = mask
tmp
```

```python
mask2 = np.hstack([
    np.zeros(196 - 147), # UNMASK
    np.ones(147),                     # MASK
])
tmp2 = mask2
tmp2
```

```python
print(tmp.shape)
```

```python
import torch
# numpy 배열을 PyTorch 텐서로 변환
tensor_a = torch.tensor(tmp)

# 텐서의 형태를 [1, 196]으로 변경
tensor_a = tensor_a.view(1, -1)

print(tensor_a.shape)  # 결과: torch.Size([1, 196])
```

```python
import torch

# 예시 tensor_a와 tensor_b
tensor_a = torch.randn(1, 196)
tensor_b = torch.randn(1, 196)

# torch.stack을 사용하여 두 텐서를 쌓기
result = torch.stack((tensor_a, tensor_b), dim=0)

print(result.shape)  # 결과: torch.Size([2, 196])
```

```python
import torch

# 예시 tensor_a와 tensor_b
tensor_a = torch.randn(1, 196)
tensor_b = torch.randn(1, 196)

# torch.cat을 사용하여 두 텐서를 연결하기
result = torch.cat((tensor_a, tensor_b), dim=0)

print(result.shape)  # 결과: torch.Size([2, 196])
```
